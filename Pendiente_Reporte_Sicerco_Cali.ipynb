```python
#Aqui carga la base de inspecciín cali, Pasto,
# Abre un cuadro de diálogo para seleccionar y subir archivos
# Aqui carga el archivo Base Reporte_Dictamenes de Sicerco
uploaded = files.upload()

# Opcional: ver los nombres de los archivos subidos
for filename_2 in uploaded.keys():
    print(f'Se subió el archivo: {filename_2}')
```

```python
# imprimir el nombre del archivo
print(filename_2)
```

```python
import pandas as pd

# Reemplaza con el nombre del archivo subido
df_BaseInspecc = pd.read_excel(filename_2)

# Ver las primeras filas
df_BaseInspecc.head()
```

```python
len(df_BaseInspecc)
```

```python
# Actas de Inspección pendientes de cargue a Sicerco:

# Asegurar que ambas columnas sean del mismo tipo (recomendado)
df_BaseInspecc["No. INFORME"] = df_BaseInspecc["No. INFORME"].astype(str)
df_filtrado["Número de dictamen"] = df_filtrado["Número de dictamen"].astype(str)

# Filtrar los registros que NO están en el otro dataframe Base Sicerco
df_resultado = df_BaseInspecc[
    ~df_BaseInspecc["No. INFORME"].isin(df_filtrado["Número de dictamen"])
]

# Ver resultado
print(df_resultado)

# Exportar a Excel
df_resultado.to_excel("df_resultado.xlsx", index=False)
```

```python
# Descragar el Excel de la pendientes a Sicerco:
files.download('df_resultado.xlsx')
```
